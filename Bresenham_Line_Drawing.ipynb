```python
import matplotlib.pyplot as plt
```

```python
def bresenham(x0,y0,x1,y1):

  xcoordinates =[]
  ycoordinates=[]
  dx=x1-x0
  dy=y1-y0
  x=x0
  y=y0
  p=2*dy-dx

  while(x<x1):

     if(p>=0):
      y=y+1
      p=p+2*(dy-dx)
     else:
        p=p+2*dy
     x=x+1

     print("x: ",x, end=" ")
     print("y: ",y, end="\n")

     xcoordinates.append(x)
     ycoordinates.append(y)

  plt.plot(xcoordinates,ycoordinates, marker="o",markersize=3,markerfacecolor="")
  plt.show()
```

```python
bresenham(5,5,10,10)
```
